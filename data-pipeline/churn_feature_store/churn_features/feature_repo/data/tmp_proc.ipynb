```python
import pandas as pd

df = pd.read_parquet('./processed_churn_data.parquet')
```

```python
train_df = df.sample(frac=0.8, random_state=123)
test_df = df.drop(train_df.index)
```

```python
train_df.to_parquet(
    "./train.parquet",
    index=False
)

test_df.to_parquet(
    "./test.parquet",
    index=False
)

print(f"Train shape: {train_df.shape}")
print(f"Test shape: {test_df.shape}")
```
